```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer

model = AutoModelForSequenceClassification.from_pretrained("AnushS/autotrain-product_categorization-51595122230", use_auth_token=True)

tokenizer = AutoTokenizer.from_pretrained("AnushS/autotrain-product_categorization-51595122230", use_auth_token=True)

inputs = tokenizer("I love AutoTrain", return_tensors="pt")

outputs = model(**inputs)
```

```python
titles = df["TITLE"].to_list()
```

```python
inputs = []
for element in titles:
    inputs.append([tokenizer(element, return_tensors="pt")])
    
```

```python
outputs = []
for elements in inputs:
    input = str(element)
    outputs.append(model(**input))
```
